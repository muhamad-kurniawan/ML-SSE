```python
import pandas as pd
import math
import itertools
import re
import os
```

```python
base_dir = 'D:/GITHUB/ML_SSE'

with open(os.path.join(base_dir, 'element_substitution/pair_correlations.txt')) as f:
    corr = f.readlines()
```

```python
corr = [x.strip('\n') for x in corr]
corr
```

```python
df = pd.DataFrame()
for l in corr:
    split = l.split(' ')
    df_temp = pd.DataFrame({'ionA':[split[0]], 'ionB':[split[1]], 'gAB':[split[2]]})
    df = pd.concat([df, df_temp])
df = df.iloc[1:].reset_index(drop=True)
df
```

```python
df['gAB_log'] = df['gAB'].apply(lambda x: math.log(float(x),10))
```

Filter substitution with gAB_log > 0.5

```python
df = df[df['gAB_log']>0.5]
df = df.reset_index(drop=True)
```

```python
#check element substitution for Fe2+

df[df['ionA']=='Fe:2+']
```

```python
#list of all unique element

unique_element = list(set(list(df['ionA'])+list(df['ionB'])))
unique_element
```

```python
#create dict contain possible substitution of element

possible_subs = {}

for n in unique_element:
    ions_temp = []
    corrs_temp = []
    filter1 = df[df['ionA']==n]
    filter2 = df[df['ionB']==n]
    ions_temp.extend(list(filter1['ionB']))
    corrs_temp.extend(list(filter1['gAB_log']))
    ions_temp.extend(list(filter2['ionA']))
    corrs_temp.extend(list(filter2['gAB_log']))
    ions_temp = [y for x, y in sorted(zip(corrs_temp, ions_temp), reverse=True)]
    corrs_temp = [x for x, y in sorted(zip(corrs_temp, ions_temp), reverse=True)]
    n = n.replace(':', '')
    ions_temp = [x.replace(':', '') for x in ions_temp]
    possible_subs[n] = [ions_temp, corrs_temp]

possible_subs
```

```python
data_to_subs = pd.read_csv(os.path.join(base_dir, 'element_substitution/data_to_sub.csv'))
data_to_subs
```

```python
def extract_ox_num(ion):
    return re.match(r"([A-Za-z]+)([0-9]+[+-])", ion, re.I).groups()[1]
def extract_elem(ion):
    return re.match(r"([A-Za-z]+)([0-9]+[+-])", ion, re.I).groups()[0]
def create_comb(dataframe, dict_elem_subs):
    n=0
    r=0
    all_combs = []
    df_ = dataframe.loc[:, ['ion_tosub1', 'ion_tosub2', 'ion_tosub3', 'ion_tosub4']]
    list_df = df_.values.tolist()
    for row in list_df:
        row_combs = []
        list_elem_subs = []
        cleaned_row = [x for x in row if str(x) != 'nan']
        for elem in cleaned_row:
            try:
                ox_number = extract_ox_num(elem)
                subs = [elem]+dict_elem_subs[elem][0]
                subs = [x for x in subs if extract_ox_num(x)==ox_number]
                if len(subs)!=0:
                    list_elem_subs.append(subs)
            except:
                pass
        for comb in itertools.product(*list_elem_subs):
            row_combs.append(list(comb))
        all_combs.append(row_combs)
    return all_combs

def change_ions(dataframe, dict_elem_subs):
    fomula_list = dataframe['formula']
    comb_list = create_comb(dataframe, dict_elem_subs)
    final_list = []
    assert len(fomula_list)==len(comb_list)
    for n in range(len(comb_list)):
        if len(comb_list[n][0])!=0:
            elements = [extract_elem(x) for x in comb_list[n][0]]
            total_elem = len(elements)
            new_formula_list = []
            for elems_to_change in comb_list[n][1:]:
                dict_elem = {}
                elem_to = [extract_elem(el) for el in elems_to_change]
                for n_el in range(len(elem_to)):
                    dict_elem[elements[n_el]] = elem_to[n_el]
                formula_changed = str(fomula_list[n])
                dict_elem = dict((re.escape(k), v) for k, v in dict_elem.items())
                pattern = re.compile("|".join(dict_elem.keys()))
                formula_changed = pattern.sub(lambda m: dict_elem[re.escape(m.group(0))], formula_changed)
                new_formula_list.append(formula_changed)
            final_list.append(new_formula_list)
        else:
            final_list.append([])
    return final_list
```

```python
formula_comb = change_ions(data_to_subs, possible_subs)
sum([len(x) for x in formula_comb])
#count the combs
```

```python
formula_comb
```

```python
df_final = pd.DataFrame()
base_formula = list(data_to_subs['formula'])
ionic_conductivity_formula = list(data_to_subs['target'])
space_group_num = list(data_to_subs['space group #'])
space_group = list(data_to_subs['space group'])
structure = list(data_to_subs['structure'])
for r in range(len(formula_comb)):
    for r2 in range(len(formula_comb[r])):
        df_final = pd.concat([df_final, pd.DataFrame({
            'base_formula':[base_formula[r]], 'space_group_num':[space_group_num[r]], 'space_group':[space_group[r]], 'structure':[structure[r]],
            'base_log_ionic_conductivity':[ionic_conductivity_formula[r]], 'new_formula':[formula_comb[r][r2]]
            })])
```

```python
df_final = df_final.reset_index(drop=True)
df_final
```

```python
df_final = df_final.drop_duplicates('new_formula')
```

```python
df_final.to_csv(os.path.join(base_dir, 'element_substitution/new_formula.csv'), index=False)
```
